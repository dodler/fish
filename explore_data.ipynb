```python
import pandas as pd
import os.path as osp
import os

import cv2
import matplotlib.pyplot as plt

import random
```

```python
train = os.listdir('train')
test = os.listdir('test')
```

```python
num_row = 3
num_cols = 2
```

```python
fig = plt.figure(figsize=(20,10))
k = 1
for j in range(1, num_cols+1):
    for i in range(1,num_row+1):
        img = cv2.imread(osp.join('train',random.choice(train)))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        fig.add_subplot(num_cols, num_row, k)
        plt.imshow(img)
        
        k+= 1
```

```python
fig = plt.figure(figsize=(20,10))
k = 1
for j in range(1, num_cols+1):
    for i in range(1,num_row+1):
        img = cv2.imread(osp.join('test',random.choice(test)))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        fig.add_subplot(num_cols, num_row, k)
        plt.imshow(img)
        
        k+= 1
```
